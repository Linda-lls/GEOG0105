#**Calculation of Visual Walkability Index (VWI)**


##**1. Calculating four sub-indicators:** Greenery, Crowdedness, Outdoor Enclosure, Visual Pavement

```python
import csv
import pandas as pd
import numpy as np
```

```python
# import initial categories of street features
segformer_street_feature = pd.read_csv("/content/drive/MyDrive/GEOG0105/Output_1/SegFormer/segformer_result_final.csv")
```

```python
print(segformer_street_feature.info())
segformer_street_feature
```

```python
# reclassify categories of street features in each GSV
segformer_street_feature['Building'] = segformer_street_feature['1']
segformer_street_feature['Sky'] = segformer_street_feature['2']
segformer_street_feature['Road'] = segformer_street_feature['6']
segformer_street_feature['Pavement'] = segformer_street_feature['11']
segformer_street_feature['Fence'] = segformer_street_feature['32']
segformer_street_feature['Tree'] = segformer_street_feature['4'] + segformer_street_feature['72']
segformer_street_feature['Obstacles'] = segformer_street_feature['12'] + segformer_street_feature['20'] + segformer_street_feature['43'] + segformer_street_feature['93']
print(segformer_street_feature.info())
segformer_street_feature
```

```python
# proportion of street features at each sample point
segformer_street_feature['T'] = segformer_street_feature.groupby(['id_x'])['Tree'].transform('sum')
segformer_street_feature['C'] = segformer_street_feature.groupby(['id_x'])['Obstacles'].transform('sum')
segformer_street_feature['B'] = segformer_street_feature.groupby(['id_x'])['Building'].transform('sum')
segformer_street_feature['P'] = segformer_street_feature.groupby(['id_x'])['Pavement'].transform('sum')
segformer_street_feature['R'] = segformer_street_feature.groupby(['id_x'])['Road'].transform('sum')
segformer_street_feature['F'] = segformer_street_feature.groupby(['id_x'])['Fence'].transform('sum')
segformer_street_feature
```

```python
segformer_street_feature.info()
```

```python
# save reclassify results of each GSV image as csv
csv_path1 = '/content/drive/MyDrive/GEOG0105/Output_1/SegFormer/reclassify_GSV.csv'
segformer_street_feature.to_csv(csv_path1, index=False)
```

```python
segformer_sum_pixel_sample_points = segformer_street_feature.groupby('id_x').agg(
    {'lon': 'first', 'lat': 'first', 'geometry': 'first', 'index_right': 'first', 'OBJECTID': 'first', 'PROVINCE': 'first', 'CAT_B': 'first',
     'WARDNO': 'first', 'WARD_ID': 'first', 'SUM_WARD_P': 'first', 'SHAPE_LENG': 'first', 'WARD_EXT': 'first', 'ALBERS_ARE': 'first',
     'SHAPE_Le_1': 'first', 'SHAPE_Area': 'first', 'T': 'first', 'C': 'first', 'B': 'first', 'P': 'first', 'R': 'first', 'F': 'first'})

segformer_sum_pixel_sample_points.reset_index(inplace=True)

print(segformer_sum_pixel_sample_points.info())
segformer_sum_pixel_sample_points
```

```python
# Four sub-indicators calculation
segformer_sum_pixel_sample_points['Gi'] = segformer_sum_pixel_sample_points['T'] / (3*1)
segformer_sum_pixel_sample_points['Ci'] = segformer_sum_pixel_sample_points['C'] / (3*1)
segformer_sum_pixel_sample_points['Si'] = (segformer_sum_pixel_sample_points['B'] + segformer_sum_pixel_sample_points['T']) / (segformer_sum_pixel_sample_points['P'] + segformer_sum_pixel_sample_points['R'] + segformer_sum_pixel_sample_points['F'])
segformer_sum_pixel_sample_points['Di'] = (segformer_sum_pixel_sample_points['P'] + segformer_sum_pixel_sample_points['F']) / segformer_sum_pixel_sample_points['R']
```

```python
print(segformer_sum_pixel_sample_points.info())
```

```python
segformer_sum_pixel_sample_points
```

```python
segformer_sum_pixel_sample_points = segformer_sum_pixel_sample_points.fillna(0)

segformer_sum_pixel_sample_points = segformer_sum_pixel_sample_points.replace([np.inf, -np.inf], 0)
```

```python
segformer_sum_pixel_sample_points.info()
```

```python
segformer_sum_pixel_sample_points
```

```python
# save sub-indicators results of each sample point as csv
csv_path2 = '/content/drive/MyDrive/GEOG0105/Output_1/SegFormer/sub-indicators_result.csv'
segformer_sum_pixel_sample_points.to_csv(csv_path2, index=False)
```

##**2. Calculation of IVW**

```python
# import sub-insicator results
segformer_sum_pixel_sample_points = pd.read_csv("/content/drive/MyDrive/GEOG0105/Output_1/SegFormer/sub-indicators_result.csv")
segformer_sum_pixel_sample_points
```

```python
Gi_ranges = [0, 0.070, 0.127, 0.175, 0.228, float('inf')]
Gi_levels = [1, 2, 3, 4, 5]
segformer_sum_pixel_sample_points['Gi_level'] = pd.cut(segformer_sum_pixel_sample_points['Gi'], bins=Gi_ranges, labels=Gi_levels, right=False)
```

```python
Ci_ranges = [0, 0.169, 0.286, 0.413, 0.574, float('inf')]
Ci_levels = [5, 4, 3, 2, 1]
segformer_sum_pixel_sample_points['Ci_level'] = pd.cut(segformer_sum_pixel_sample_points['Ci'], bins=Ci_ranges, labels=Ci_levels, right=False,  ordered=False)
```

```python
Si_ranges = [0, 0.800, 1.063, 1.213, 1.350, 1.656, 1.884, 2.210, 2.864, float('inf')]
Si_levels = [1, 2, 3, 4, 5, 4, 3, 2, 1]
segformer_sum_pixel_sample_points['Si_level'] = pd.cut(segformer_sum_pixel_sample_points['Si'], bins=Si_ranges, labels=Si_levels, right=False, ordered=False)
```

```python
Di_ranges = [0, 0.480, 0.650, 0.806, 0.953, 1.707, 2.137, 3.007, 4.000, float('inf')]
Di_levels = [1, 2, 3, 4, 5, 4, 3, 2, 1]
segformer_sum_pixel_sample_points['Di_level'] = pd.cut(segformer_sum_pixel_sample_points['Di'], bins=Di_ranges, labels=Di_levels, right=False, ordered=False)
```

```python
segformer_sum_pixel_sample_points
```

```python
segformer_sum_pixel_sample_points['IVW'] = 5 * (segformer_sum_pixel_sample_points['Gi_level'].astype(int)
 + segformer_sum_pixel_sample_points['Ci_level'].astype(int) + segformer_sum_pixel_sample_points['Si_level'].astype(int)
  + segformer_sum_pixel_sample_points['Di_level'].astype(int))
```

```python
print(segformer_sum_pixel_sample_points.info())
segformer_sum_pixel_sample_points
```

```python
# save sub-indicators results of each sample point as csv
csv_path3 = '/content/drive/MyDrive/GEOG0105/Output_1/SegFormer/IVW.csv'
segformer_sum_pixel_sample_points.to_csv(csv_path3, index=False)
```
